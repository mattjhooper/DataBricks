# Create a Crowd2Fund Investments Table from a csv export

Setup Variables

```python
catalog = "crowd2fund"
schema = "default"
volume = "rawdata/investments"
table_name = "my_investments"
path_volume = "/Volumes/" + catalog + "/" + schema + "/" + volume
path_table = catalog + "." + schema
print(path_table) # Show the complete path
print(path_volume) # Show the complete path
```

Load CSV into a DataFrame

```python
df = spark.read.csv(f"{path_volume}",
  header=True,
  inferSchema=True,
  sep=",")
```

```python
display(df)
```

Save the DataFrame to a table

```python
import re

def clean_col(col_name):
    return re.sub(r'[^A-Za-z0-9_]', '', col_name.replace(' ', '_'))

df = df.toDF(*[clean_col(c) for c in df.columns])
df.printSchema()
```

```python
df.write.mode("overwrite").saveAsTable(f"{path_table}" + "." + f"{table_name}")
```
