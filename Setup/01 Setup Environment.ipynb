# Setup Crowd2Fund Environment

## 1. Access Crowd2Fund Container

```sql
%python
storageAccount = "crowd2fund"
location = f"abfss://crowd2fund@crowd2fund.dfs.core.windows.net/"
path = (
    f"{location}landing/exports"
)
display(
    dbutils.fs.ls(path)
)
```

## 2. Create External Location

```sql
DROP EXTERNAL LOCATION IF EXISTS crowd2fund_ext_dl;
CREATE EXTERNAL LOCATION IF NOT EXISTS crowd2fund_ext_dl
    URL 'abfss://crowd2fund@crowd2fund.dfs.core.windows.net/'
    WITH (STORAGE CREDENTIAL crowd2fund_ext_sc)
    COMMENT 'External Location For the crowd2fund data extracts'
```

## 3. Create the catalog - Crowd2Fund

```sql
CREATE CATALOG IF NOT EXISTS crowd2fund
      MANAGED LOCATION 'abfss://crowd2fund@crowd2fund.dfs.core.windows.net/'
      COMMENT 'This is the catalog for the Crowd2Fund Data Lakehouse' ;
```

## 4. Create the Schemas

```sql
USE CATALOG crowd2fund;

CREATE SCHEMA IF NOT EXISTS landing
     MANAGED LOCATION 'abfss://crowd2fund@crowd2fund.dfs.core.windows.net/landing';
CREATE SCHEMA IF NOT EXISTS bronze
     MANAGED LOCATION 'abfss://crowd2fund@crowd2fund.dfs.core.windows.net/bronze';  
CREATE SCHEMA IF NOT EXISTS silver
     MANAGED LOCATION 'abfss://crowd2fund@crowd2fund.dfs.core.windows.net/silver';  
CREATE SCHEMA IF NOT EXISTS gold
     MANAGED LOCATION 'abfss://crowd2fund@crowd2fund.dfs.core.windows.net/gold';  
```

```sql
SHOW SCHEMAS;
```

## 5. Create Volume

```sql
USE CATALOG crowd2fund;
USE SCHEMA landing;

CREATE EXTERNAL VOLUME IF NOT EXISTS exports
    LOCATION 'abfss://crowd2fund@crowd2fund.dfs.core.windows.net/landing/exports/';
```
