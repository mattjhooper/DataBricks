# Create Bronze Views

## Add Investments View

```sql
CREATE OR REPLACE VIEW crowd2fund.bronze.v_investments AS
SELECT *,
      _metadata.file_path AS file_path 
  FROM read_files('/Volumes/crowd2fund/landing/exports/investments',
                  format => 'csv',
                  delimiter => ',',
                  header => true);
```

```sql
SELECT * FROM crowd2fund.bronze.v_investments;
```

## Add Transactions View

```sql
CREATE OR REPLACE VIEW crowd2fund.bronze.v_transactions AS
SELECT *,
      _metadata.file_path AS file_path 
  FROM read_files('/Volumes/crowd2fund/landing/exports/transactions',
                  format => 'csv',
                  delimiter => ',',
                  header => true);
```
